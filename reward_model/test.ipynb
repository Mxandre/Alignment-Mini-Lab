```python
from datasets import load_dataset
```

```python
raw_ds = load_dataset("HuggingFaceH4/ultrafeedback_binarized")
```

```python
raw_ds["train_prefs"]["prompt"]
```

```python
from transformers import AutoTokenizer
tokenizer = AutoTokenizer.from_pretrained("Qwen/Qwen2.5-7B")
```

```python
lengths = []
for i in range(len(raw_ds["train_prefs"])):
    content = raw_ds["train_prefs"][i]["chosen"][1]["content"] + raw_ds["train_prefs"][i]["prompt"] # 获取模型回复内容
    tokens = tokenizer.encode(content)
    lengths.append(len(tokens))

print(f"平均长度: {sum(lengths)/len(raw_ds['train_prefs'])}")
print(f"最大长度: {max(lengths)}")
print(f"超过 1024 的比例: {len([l for l in lengths if l > 1024]) / 10}%")
```
